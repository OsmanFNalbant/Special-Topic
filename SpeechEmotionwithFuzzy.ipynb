```python
import pandas as pd
import numpy as np
import os
import librosa

data_dir = "C:\\AudioWAV\\"

data = [f for f in os.listdir(data_dir) if f.endswith('.wav')]
```

```python
def extract_features(data):
    features = {}
    for file in data:
        print("Proccesing file: ", file)
        y, sr = librosa.load(data_dir + file)
        
        # Preprocessing
        yt, index= librosa.effects.trim(y)
        
        # features
        
        mfccs = librosa.feature.mfcc(y=yt, sr=sr, n_mfcc=5)
        pitch = librosa.effects.harmonic(y=yt)
        zcr = librosa.feature.zero_crossing_rate(yt)
        rmse = librosa.feature.rms(y=yt)
        speaking_rate = len(yt) / sr
        
        
        features[file] = {
            'mfcc_1': mfccs[0],
            'mfcc_2': mfccs[1],
            'mfcc_3': mfccs[2],
            'mfcc_4': mfccs[3],
            'mfcc_5': mfccs[4],
            'pitch': pitch,
            'zcr': zcr,
            'rmse': rmse,
            'speaking_rate': speaking_rate
        }
        print("Done")
    return features

extracted_features = extract_features(data)

feature_df = pd.DataFrame.from_dict(extracted_features, orient='index')

feature_df.head()
```

```python
# Mean of each feature

meaned_features = feature_df.copy()

def mean_features(data):
    return np.mean(data)

meaned_features['mfcc_1'] = meaned_features['mfcc_1'].apply(mean_features)
meaned_features['mfcc_2'] = meaned_features['mfcc_2'].apply(mean_features)
meaned_features['mfcc_3'] = meaned_features['mfcc_3'].apply(mean_features)
meaned_features['mfcc_4'] = meaned_features['mfcc_4'].apply(mean_features)
meaned_features['mfcc_5'] = meaned_features['mfcc_5'].apply(mean_features)
meaned_features['pitch'] = meaned_features['pitch'].apply(mean_features)
meaned_features['zcr'] = meaned_features['zcr'].apply(mean_features)
meaned_features['rmse'] = meaned_features['rmse'].apply(mean_features)

meaned_features.head()
```

```python
## Save the features to a csv file
feature_df.to_csv('features.csv')
meaned_features.to_csv('meaned_features.csv')
```

```python
# Normalize the features
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

normalized_features = meaned_features.copy()

normalized_features['mfcc_1'] = scaler.fit_transform(normalized_features['mfcc_1'].values.reshape(-1, 1))
normalized_features['mfcc_2'] = scaler.fit_transform(normalized_features['mfcc_2'].values.reshape(-1, 1))
normalized_features['mfcc_3'] = scaler.fit_transform(normalized_features['mfcc_3'].values.reshape(-1, 1))
normalized_features['mfcc_4'] = scaler.fit_transform(normalized_features['mfcc_4'].values.reshape(-1, 1))
normalized_features['mfcc_5'] = scaler.fit_transform(normalized_features['mfcc_5'].values.reshape(-1, 1))
normalized_features['pitch'] = scaler.fit_transform(normalized_features['pitch'].values.reshape(-1, 1))
normalized_features['zcr'] = scaler.fit_transform(normalized_features['zcr'].values.reshape(-1, 1))
normalized_features['rmse'] = scaler.fit_transform(normalized_features['rmse'].values.reshape(-1, 1))
normalized_features['speaking_rate'] = scaler.fit_transform(normalized_features['speaking_rate'].values.reshape(-1, 1))

normalized_features.head()
```

```python
def calculate_feature_ranges(df):
    ranges = {}
    for emotion in df['label'].unique():
        emotion_subset = df[df['label'] == emotion]
        ranges[emotion] = {}
        for feature in ['zcr', 'speaking_rate', 'pitch', 'mfcc_1', 'mfcc_2', 'mfcc_3', 'mfcc_4', 'mfcc_5', 'rmse']:
            # Çeyreklik metoduyla aralık hesaplama
            Q1 = emotion_subset[feature].quantile(0.25)
            Q3 = emotion_subset[feature].quantile(0.75)
            IQR = Q3 - Q1
            
            # Genişletilmiş aralık (1.5 IQR)
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            ranges[emotion][feature] = {
                'min': lower_bound,
                'max': upper_bound,
                'mean': emotion_subset[feature].mean(),
                'median': emotion_subset[feature].median()
            }
    return ranges
```

```python
emotion_feature_ranges = calculate_feature_ranges(encoded_features)

for emotion, features in emotion_feature_ranges.items():
    print(f"\n\n{emotion} Emotion Feature Ranges: ")
    for feature, stats in features.items():
        print(f"\n{feature} Stats: ")
        print(stats)
```

```python
def create_membership_ranges(df):
    membership_ranges = {}
    for emotion in df['label'].unique():
        emotion_subset = df[df['label'] == emotion]
        membership_ranges[emotion] = {}
        
        for feature in ['zcr', 'speaking_rate', 'pitch', 'mfcc_1', 'mfcc_2', 'mfcc_3', 'mfcc_4', 'mfcc_5', 'rmse']:
            # Üçgen üyelik fonksiyonu için noktalar
            Q1 = emotion_subset[feature].quantile(0.25)
            Q2 = emotion_subset[feature].median()
            Q3 = emotion_subset[feature].quantile(0.75)
            
            membership_ranges[emotion][feature] = [
                max(df[feature].min(), Q1 - (Q3-Q1)),  # Alt sınır
                Q2,  # Orta nokta
                min(df[feature].max(), Q3 + (Q3-Q1))  # Üst sınır
            ]
    
    return membership_ranges
```

```python
membership_ranges= create_membership_ranges(encoded_features)

for emotion, features in membership_ranges.items():
    print(f"\n\n{emotion} Emotion Membership Ranges: ")
    for feature, ranges in features.items():
        print(f"\n{feature} Ranges: ")
        print(ranges)
```

```python
import numpy as np
import pandas as pd
import skfuzzy as fuzz
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import (
    classification_report, 
    confusion_matrix, 
    accuracy_score
)
from keras.callbacks import ModelCheckpoint, EarlyStopping,ReduceLROnPlateau
import tensorflow.keras.layers as L
from tensorflow.keras.layers import Dropout
from matplotlib import pyplot as plt
import seaborn as sns

class HybridEmotionRecognition:
    def __init__(self, data):
        self.data = data
        self.emotions = ['ANG', 'DIS', 'FEA', 'HAP', 'NEU', 'SAD']
    
    def create_fuzzy_membership_functions(self):
        membership_points = {
            'ANG': {
                'zcr': create_membership_ranges(self.data)['ANG']['zcr'],
                'mfcc_1': create_membership_ranges(self.data)['ANG']['mfcc_1'],
                'mfcc_2': create_membership_ranges(self.data)['ANG']['mfcc_2'],
                'mfcc_3': create_membership_ranges(self.data)['ANG']['mfcc_3'],
                'mfcc_4': create_membership_ranges(self.data)['ANG']['mfcc_4'],
                'mfcc_5': create_membership_ranges(self.data)['ANG']['mfcc_5'],
                'rmse': create_membership_ranges(self.data)['ANG']['rmse'],
                'speaking_rate': create_membership_ranges(self.data)['ANG']['speaking_rate'],
                'pitch': create_membership_ranges(self.data)['ANG']['pitch']
            },
            
            'DIS': {'zcr': create_membership_ranges(self.data)['DIS']['zcr'],
                    'mfcc_1': create_membership_ranges(self.data)['DIS']['mfcc_1'],
                    'mfcc_2': create_membership_ranges(self.data)['DIS']['mfcc_2'],
                    'mfcc_3': create_membership_ranges(self.data)['DIS']['mfcc_3'],
                    'mfcc_4': create_membership_ranges(self.data)['DIS']['mfcc_4'],
                    'mfcc_5': create_membership_ranges(self.data)['DIS']['mfcc_5'],
                    'rmse': create_membership_ranges(self.data)['DIS']['rmse'],
                    'speaking_rate': create_membership_ranges(self.data)['DIS']['speaking_rate'],
                    'pitch': create_membership_ranges(self.data)['DIS']['pitch']
                    },
            
            'FEA': {'zcr': create_membership_ranges(self.data)['FEA']['zcr'],
                    'mfcc_1': create_membership_ranges(self.data)['FEA']['mfcc_1'],
                    'mfcc_2': create_membership_ranges(self.data)['FEA']['mfcc_2'],
                    'mfcc_3': create_membership_ranges(self.data)['FEA']['mfcc_3'],
                    'mfcc_4': create_membership_ranges(self.data)['FEA']['mfcc_4'],
                    'mfcc_5': create_membership_ranges(self.data)['FEA']['mfcc_5'],
                    'rmse': create_membership_ranges(self.data)['FEA']['rmse'],
                    'speaking_rate': create_membership_ranges(self.data)['FEA']['speaking_rate'],
                    'pitch': create_membership_ranges(self.data)['FEA']['pitch']
                    },
            
            'HAP': {'zcr': create_membership_ranges(self.data)['HAP']['zcr'],
                    'mfcc_1': create_membership_ranges(self.data)['HAP']['mfcc_1'],
                    'mfcc_2': create_membership_ranges(self.data)['HAP']['mfcc_2'],
                    'mfcc_3': create_membership_ranges(self.data)['HAP']['mfcc_3'],
                    'mfcc_4': create_membership_ranges(self.data)['HAP']['mfcc_4'],
                    'mfcc_5': create_membership_ranges(self.data)['HAP']['mfcc_5'],
                    'rmse': create_membership_ranges(self.data)['HAP']['rmse'],
                    'speaking_rate': create_membership_ranges(self.data)['HAP']['speaking_rate'],
                    'pitch': create_membership_ranges(self.data)['HAP']['pitch']
                    },
            
            'NEU': {'zcr': create_membership_ranges(self.data)['NEU']['zcr'],
                    'mfcc_1': create_membership_ranges(self.data)['NEU']['mfcc_1'],
                    'mfcc_2': create_membership_ranges(self.data)['NEU']['mfcc_2'],
                    'mfcc_3': create_membership_ranges(self.data)['NEU']['mfcc_3'],
                    'mfcc_4': create_membership_ranges(self.data)['NEU']['mfcc_4'],
                    'mfcc_5': create_membership_ranges(self.data)['NEU']['mfcc_5'],
                    'rmse': create_membership_ranges(self.data)['NEU']['rmse'],
                    'speaking_rate': create_membership_ranges(self.data)['NEU']['speaking_rate'],
                    'pitch': create_membership_ranges(self.data)['NEU']['pitch']
                    },
            
            'SAD': {'zcr': create_membership_ranges(self.data)['SAD']['zcr'],
                    'mfcc_1': create_membership_ranges(self.data)['SAD']['mfcc_1'],
                    'mfcc_2': create_membership_ranges(self.data)['SAD']['mfcc_2'],
                    'mfcc_3': create_membership_ranges(self.data)['SAD']['mfcc_3'],
                    'mfcc_4': create_membership_ranges(self.data)['SAD']['mfcc_4'],
                    'mfcc_5': create_membership_ranges(self.data)['SAD']['mfcc_5'],
                    'rmse': create_membership_ranges(self.data)['SAD']['rmse'],
                    'speaking_rate': create_membership_ranges(self.data)['SAD']['speaking_rate'],
                    'pitch': create_membership_ranges(self.data)['SAD']['pitch']
                    }
        }
        
        return membership_points
    
    def calculate_fuzzy_membership(self, sample, emotion_points):
        fuzzy_memberships = {}
        
        for emotion, feature_points in emotion_points.items():
            emotion_degree = 1.0
            
            for feature, points in feature_points.items():
                # Üçgensel üyelik fonksiyonu 
                membership_values = fuzz.trimf(
                    np.array([sample[feature]]), 
                    points
                )[0]
                
                emotion_degree *= membership_values
            
            fuzzy_memberships[emotion] = emotion_degree
        
        return fuzzy_memberships
    
    def prepare_hybrid_features(self, X, emotion_points):
        hybrid_features = []
        
        for _, sample in X.iterrows():
            # Özgün özellikleri al
            original_features = sample[['zcr', 'mfcc_1', 'mfcc_2','mfcc_3', 'mfcc_4','mfcc_5', 'rmse', 'speaking_rate', 'pitch']].values
            
            # Bulanık üyelik derecelerini hesapla
            fuzzy_degrees = self.calculate_fuzzy_membership(sample, emotion_points)
            fuzzy_features = list(fuzzy_degrees.values())
            
            # Özgün ve bulanık özellikleri birleştir
            hybrid_feature = np.concatenate([original_features, fuzzy_features])
            hybrid_features.append(hybrid_feature)
        
        return np.array(hybrid_features)
    
    def build_hybrid_cnn_model(self, input_shape, num_classes):
        model_checkpoint = ModelCheckpoint('best_model.weights.h5', monitor='val_accuracy', save_best_only=True, save_weights_only=True)
        early_stop=EarlyStopping(monitor='val_accuracy',mode='auto', patience=7, restore_best_weights=True)
        lr_reduction=ReduceLROnPlateau(monitor='val_accuracy', patience=5, verbose=1, factor=0.9, min_lr=0.001)
        
        
        
        model = tf.keras.Sequential([
            L.Conv1D(128,kernel_size=5, strides=1,padding='same', activation='relu',input_shape=(input_shape[1],1)),
            L.BatchNormalization(),
            L.MaxPool1D(pool_size=5,strides=2,padding='same'),

            L.Conv1D(64,kernel_size=5,strides=1,padding='same',activation='relu'),
            L.BatchNormalization(),
            L.MaxPool1D(pool_size=5,strides=2,padding='same'),

            L.Conv1D(256,kernel_size=3,strides=1,padding='same',activation='relu'),
            L.BatchNormalization(),
            L.MaxPool1D(pool_size=5,strides=2,padding='same'),

            L.Conv1D(128,kernel_size=3,strides=1,padding='same',activation='relu'),
            L.BatchNormalization(),
            L.MaxPool1D(pool_size=3,strides=2,padding='same'),
            Dropout(0.2),

            L.Conv1D(64,kernel_size=3,strides=1,padding='same',activation='relu'),
            L.BatchNormalization(),
            L.MaxPool1D(pool_size=3,strides=2,padding='same'),
            

            L.Flatten(),
            L.Dense(64,activation='relu'),
            L.BatchNormalization(),
            L.Dense(num_classes,activation='softmax')
                ])
        optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
        model.compile(optimizer=optimizer,loss='categorical_crossentropy',metrics=['accuracy'])
        model.summary()
        return model, model_checkpoint, early_stop, lr_reduction
    
    def train_hybrid_model(self, X, y):
        
        # Verileri eğitim ve test setlerine ayır
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.25, random_state=42,shuffle=True
        )
        print("xtrain, xtest,yteain,ytest: ",X_train.shape, X_test.shape, y_train.shape, y_test.shape)
        
        
        # Etiketleri kodla
        label_encoder = LabelEncoder()
        y_train_encoded = label_encoder.fit_transform(y_train)
        y_test_encoded = label_encoder.transform(y_test)
        
        # One-hot encoding
        y_train_onehot = tf.keras.utils.to_categorical(y_train_encoded)
        y_test_onehot = tf.keras.utils.to_categorical(y_test_encoded)
        
        print("y tarin test onehot: ",y_train_onehot.shape, y_test_onehot.shape)
        
        
        # Özellikleri ölçeklendir
        scaler = StandardScaler()
        X_train_scaled = scaler.fit_transform(X_train)
        X_test_scaled = scaler.transform(X_test)
        
        print("without fuzzy X_train, X_test :",X_train_scaled.shape, X_test_scaled.shape)
        
        
        
        # Bulanık üyelik noktalarını oluştur
        emotion_points = self.create_fuzzy_membership_functions()
        print("Emotion points: ",emotion_points)
        
        # Hibrit özellikleri hazırla
        X_train_hybrid = self.prepare_hybrid_features(X_train, emotion_points)
        X_test_hybrid = self.prepare_hybrid_features(X_test, emotion_points)
        print("X_train_hybrid, X_test_hybrid: ",X_train_hybrid.shape, X_test_hybrid.shape)
        
        # Giriş şeklini hazırla
        input_shape = (X_train_hybrid.shape[0],X_train_hybrid.shape[1], 1)
        num_classes = len(np.unique(y))
        
        # Modeli oluştur
        model,model_checkpoint,lr_reduction, early_stop= self.build_hybrid_cnn_model(input_shape, num_classes)
        
        # Modeli şekillendir
        X_train_reshaped = X_train_hybrid.reshape(-1, X_train_hybrid.shape[1], 1)
        X_test_reshaped = X_test_hybrid.reshape(-1, X_test_hybrid.shape[1], 1)
        
        print("xtrainreshaped,ytrain, xtestreshaped, ytest: ",X_train_reshaped.shape, y_train_onehot.shape,X_test_reshaped.shape, y_test_onehot.shape)
        
        
        # Modeli eğit
        history=model.fit(X_train_reshaped, y_train_onehot, epochs=120, validation_data=(X_test_reshaped, y_test_onehot), batch_size=64,callbacks=[early_stop,model_checkpoint])
        
        return model, history, X_test_reshaped, y_test, label_encoder
    
    def evaluate_model(self, model, X_test, y_test, label_encoder):
        # Test verilerini tahmin etme
        y_pred_proba = model.predict(X_test)
        y_pred = np.argmax(y_pred_proba, axis=1)
        
        # Gerçek etiketleri decode etme
        y_true = label_encoder.transform(y_test)
        
        # Performans metrikleri
        print("\n--- Model Performance ---")
        print("Accuracy:", accuracy_score(y_true, y_pred))
        
        print("\nClassification Report:")
        print(classification_report(y_true, y_pred, 
                                    target_names=self.emotions))
        
        # Karışıklık matrisi
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(10,7))
        sns.heatmap(cm, annot=True, fmt='d', linecolor='white', cmap='Blues',
                    xticklabels=self.emotions, 
                    yticklabels=self.emotions)
        plt.title('Confusion Matrix')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.tight_layout()
        plt.show()
        
        return y_pred
    
```

```python
def main():
        # Örnek veri yükleme (gerçek veri setinizle değiştirin)
        data = pd.read_csv("normalized_features.csv")
        
        data = data.drop('file', axis=1)
        
        # Hibrit modeli oluştur ve eğit
        hybrid_emotion_model = HybridEmotionRecognition(data)
        
        model, training_history, X_test, y_test, label_encoder = hybrid_emotion_model.train_hybrid_model(
            data.drop('label', axis=1), 
            data['label']
        )
        
        # Eğitim geçmişini görselleştir
        plt.figure(figsize=(12, 4))
        
        plt.subplot(1, 2, 1)
        plt.plot(training_history.history['accuracy'], label='Training Accuracy')
        plt.plot(training_history.history['val_accuracy'], label='Validation Accuracy')
        plt.title('Model Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        
        plt.subplot(1, 2, 2)
        plt.plot(training_history.history['loss'], label='Training Loss')
        plt.plot(training_history.history['val_loss'], label='Validation Loss')
        plt.title('Model Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        
        plt.tight_layout()
        plt.show()
        
        # Modeli değerlendir
        hybrid_emotion_model.evaluate_model(
            model, X_test, y_test, label_encoder
        )
        
        
        # Modeli değerlendir
        print("Training done..")


main()
```
